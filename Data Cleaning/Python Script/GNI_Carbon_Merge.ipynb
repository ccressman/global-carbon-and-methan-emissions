```python
#import dependencies
import pandas as pd
```

```python
#read in raw GNI data and cleaned carbon dataset
raw_GNI = pd.read_excel('../Raw Data/Excel/GNI_by_country_worldbank.xls')
carbon_dataset = pd.read_excel('../Cleaned Datasets/Carbon_justcountries.xlsx')
```

```python
df = raw_GNI.drop([0,1]).reset_index(drop=True) #drop first two rows and reset index
df
```

```python
df.columns = df.iloc[0]#set first row to header
df = df[1:]#reset df to not include header row as row
df
```

```python
print(df.columns)
```

```python
GNI_data = pd.DataFrame(df, columns=['Country Code','Country Name',2021.0])
GNI_data = pd.DataFrame(df, columns=['Country Code',2021.0])
```

```python
GNI_data
```

```python
GNI_data.rename(columns = {2021.0: '2021_GNI'}, inplace=True)
GNI_data
```

```python
merged_data = pd.merge(carbon_dataset,GNI_data, how="left", on=['Country Code','Country Code'])
```

```python
#print(carbon_dataset.columns)
merged_data
```

```python
merged_data.to_csv('../../resources/carbon_and_GNI.csv',sep=',')
merged_data.to_json('../../resources/carbon_and_GNI.json')
```
